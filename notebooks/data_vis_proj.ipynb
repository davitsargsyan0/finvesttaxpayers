```python
import pandas as pd
import re
```

```python
data = pd.read_csv('pek_data_2024.11.csv')

data_cleaned = data.dropna(subset=['Գործունեության տեսակ']).reset_index()
```

```python
data_cleaned['Գործունեության տեսակ'] = data_cleaned['Գործունեության տեսակ'].apply(lambda x: re.sub(r'[^0-9.]', '', x))
```

```python
data_cleaned = data_cleaned.loc[data_cleaned['Կարգավիճակ'].isin(['Գործող', 'Նոր գրանցված'])].reset_index(drop=True)
```

```python
data_cleaned['Ընդանուր Գործունեության տեսակ'] = data_cleaned['Գործունեության տեսակ'].apply(lambda x: x.split('.')[0])
```

```python
data_cleaned['Ընդանուր Գործունեության տեսակ 1'] = data_cleaned['Գործունեության տեսակ'].apply(lambda x: '.'.join(x.split('.')[:2]) if len(x.split('.')) > 1 else x)
```

```python
data_cleaned
```

```python
general_activity_counts = data_cleaned.groupby('Ընդանուր Գործունեության տեսակ').size().reset_index(name='Count')

general_activity_counts = general_activity_counts.sort_values(by='Count', ascending=False)

general_activity_counts
```

```python
data_spheres = pd.read_excel('Data_spheres.xlsx')
data_spheres
```

```python
merged_data = general_activity_counts.merge(data_spheres, left_on='Ընդանուր Գործունեության տեսակ', right_on='Code', how='left')

merged_data.loc[merged_data['Code'] == '01']
```
