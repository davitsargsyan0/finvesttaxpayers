```python
import pandas as pd
```

```python
df = pd.read_csv('../data/pek_data_2024.11.csv')
df_inv = pd.read_csv('../data/cleaned_eng_investments.csv')
```

```python
df_inv
```

```python
df_inv['code'] = df_inv['code'].astype(str)
```

```python
columns_to_sum = [
    '2014_total', '2014_direct', '2015_total', '2015_direct',
    '2016_total', '2016_direct', '2017_total', '2017_direct',
    '2018_total', '2018_direct', '2019_total', '2019_direct',
    '2020_total', '2020_direct', '2021_total', '2021_direct',
    '2022_total', '2022_direct', '2023_totoal', '2023_direct'
]

grouped_df = df_inv.groupby('type_of_activity').agg(
    {
        'code': 'first',  # or use 'list' to get all codes per activity
        '2014_total': 'sum',
        '2014_direct': 'sum',
        '2015_total': 'sum',
        '2015_direct': 'sum',
        '2016_total': 'sum',
        '2016_direct': 'sum',
        '2017_total': 'sum',
        '2017_direct': 'sum',
        '2018_total': 'sum',
        '2018_direct': 'sum',
        '2019_total': 'sum',
        '2019_direct': 'sum',
        '2020_total': 'sum',
        '2020_direct': 'sum',
        '2021_total': 'sum',
        '2021_direct': 'sum',
        '2022_total': 'sum',
        '2022_direct': 'sum',
        '2023_totoal': 'sum',
        '2023_direct': 'sum'
    }
).reset_index()

grouped_df
```

```python
sector_codes = list(grouped_df['code'])
```

```python
df.head()
```

```python
df = df.dropna(subset=['Գործունեության տեսակ'])
```

```python
df['sector_code'] = df['Գործունեության տեսակ'].str.extract(r'^[A-Z](\d+)')[0].str.lstrip('0')
```

```python
df['sector_code'].unique()
```

```python
df['sector_code'] = df['sector_code'].astype(str)
```

```python
df['Կարգավիճակ'].unique()
```

```python
df = df.loc[df['Կարգավիճակ'].isin(['Գործող', 'Նոր գրանցված'])].reset_index(drop=True)
```

```python
df = df.loc[df['sector_code'].isin(sector_codes)].reset_index(drop=True)
```

```python
df['Գրանցման ամսաթիվ'] = pd.to_datetime(df['Գրանցման ամսաթիվ'])
df['registration_year'] = df['Գրանցման ամսաթիվ'].dt.year
```

```python
df.to_csv('../data/taxpayers_prep_data.csv')
```
