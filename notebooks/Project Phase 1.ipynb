## Lat Long extraction

```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

geolocator = Nominatim(user_agent="geoapi")

geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

def get_lat_long(location, index, total):
    try:
        loc = geocode(location)
        if loc:
            lat, long = loc.latitude, loc.longitude
        else:
            lat, long = None, None
        print(f"Processed {index + 1}/{total}: {location} -> ({lat}, {long})")
        return pd.Series([lat, long])
    except Exception as e:
        print(f"Error for location '{location}': {e}")
        return pd.Series([None, None])

total_locations = len(df_test)

df[['latitude', 'longitude']] = df.apply(lambda row: get_lat_long(row['Իրավաբանական հասցե'], row.name, total_locations), axis=1)

print("\nFinal DataFrame with coordinates:")
print(df)
```
